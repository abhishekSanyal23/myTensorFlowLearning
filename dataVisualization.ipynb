```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
%matplotlib inline
```

```python
X = np.arange(0, 10)
```

```python
Y = X**2
```

```python
plt.plot(X,Y)
```

```python
plt.plot(X,Y,'*')
```

```python
plt.plot(X, Y, 'red')
```

```python
plt.plot(X, Y, 'o--')
```

```python
plt.plot(X, Y, '--')
```

```python
plt.plot(X, Y, 'r--')
```

```python
plt.plot(X, Y, 'o--')
plt.xlim(0,4)
plt.ylim(0,10)
plt.title('Sample Math plot library')
plt.xlabel('X-axis')
plt.ylabel('Y-label')
```
